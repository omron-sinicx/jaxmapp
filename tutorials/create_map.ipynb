Script for creating maps and storing them in `data_example/map_data`

```python
import numpy as np
import matplotlib.pyplot as plt

image = np.zeros((160, 160))
image[:2, :] = 1
image[-2:, :] = 1
image[:, :2:] = 1
image[:, -2:] = 1

image[50:110, 50:110] = 1

np.savetxt("../data_example/map_data/0000.txt", image, fmt="%d")
plt.imshow(1 - image.T, cmap="gray")
```

```python

image = np.zeros((160, 160))
image[:2, :] = 1
image[-2:, :] = 1
image[:, :2:] = 1
image[:, -2:] = 1
image[40:50, 40:50] = 1
image[40:50, 110:120] = 1
image[110:120, 40:50] = 1
image[110:120, 110:120] = 1
np.savetxt("../data_example/map_data/0001.txt", image, fmt="%d")
plt.imshow(1 - image.T, cmap="gray")
```

```python
image = np.zeros((160, 160))
image[:2, :] = 1
image[-2:, :] = 1
image[:, :2:] = 1
image[:, -2:] = 1

image[0:20, 50:100] = 1
image[60:80, 50:100] = 1
image[140:160, 50:100] = 1

np.savetxt("../data_example/map_data/0002.txt", image, fmt="%d")
plt.imshow(1 - image.T, cmap="gray")
```

```python
image = np.zeros((160, 160))
image[:2, :] = 1
image[-2:, :] = 1
image[:, :2:] = 1
image[:, -2:] = 1
for i in range(20):
    image[35 + i, (60 + i):(100 - i)] = 1
    image[125 - i, (60 + i):(100 - i)] = 1
    image[(60 + i):(100 - i), 35 + i] = 1
    image[(60 + i):(100 - i), 125 - i] = 1
np.savetxt("../data_example/map_data/0003.txt", image, fmt="%d")
plt.imshow(1 - image.T, cmap="gray")
```
